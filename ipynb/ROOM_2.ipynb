```python
kata = requests.get("https://raw.githubusercontent.com/Fauzan-Kamil/piton-data/main/rangga.txt").text
kata
```

```python
kata = kata.lower()
kata
```

```python
kata = re.sub(r"[-()\"#/@;:<>{}`+=~|.!?,]", "", kata)
kata
```

```python
token = word_tokenize(kata)
print(token)
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
stopword_remover_factory = StopWordRemoverFactory()
stemmer_factory = StemmerFactory()
ina_stemmer = stemmer_factory.create_stemmer()
ina_stopword = stopword_remover_factory.create_stop_word_remover()

print(stopword_remover_factory.get_stop_words())
```

```python
text_from_token = ' '.join(token)
print(text_from_token)

text_from_token_no_stop_words = ina_stopword.remove(text_from_token)
```

```python
stemmed_kata = ina_stemmer.stem(text_from_token_no_stop_words)
stemmed_kata
```

```python
token_untuk_kata = word_tokenize(stemmed_kata)
print(token_untuk_kata)
```

```python
jumlah_ina = {}
for kata in token_untuk_kata:
  if kata not in jumlah_ina:
    jumlah_ina[kata] = 1
  else:
    jumlah_ina[kata] += 1
```

```python
print("Jumlah Bag Of Words: ",jumlah_ina)
```
